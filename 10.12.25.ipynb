```python
# print_row prints `count` stars on the same line using recursion and nested ifs
def print_row(count, cur=1):
    if cur <= count:
        print('*', end='')           # print one star
        # nested if: decide whether to print more stars in this row
        if cur < count:
            print_row(count, cur + 1)
        # when cur == count we fall out and return to caller

# print_pattern prints rows from 1..rows using recursion and nested ifs
def print_pattern(rows, r=1):
    if r <= rows:
        print_row(r)                 # print r stars
        print()                      # newline after row
        # nested if: continue to next row if needed
        if r < rows:
            print_pattern(rows, r + 1)

# Example usage:
print_pattern(5)
```
