```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mean_battery_life = 24
std_deviation = 2

battery_hours = np.linspace(10, 38, 1000)

pdf_values = stats.norm.pdf(battery_hours, mean_battery_life, std_deviation)

plt.figure(figsize=(10, 6))
plt.plot(battery_hours, pdf_values, 'b-', linewidth = 3, label='PDF of Battery Life')
plt.fill_between(battery_hours, pdf_values, alpha = 0.3, color = 'cyan')

plt.xlabel('Battery Life (hours)', fontsize = 12)
plt.ylabel('Probability Density', fontsize = 12)
plt.title('Probability Density Function (PDF) of Smartphone Battery Life', fontsize = 14)

plt.axvline(mean_battery_life, color = 'red', linestyle = '--',
            label = f'Mean: {mean_battery_life} hours')
plt.axvline(mean_battery_life - std_deviation, color = 'purple', linestyle=':',
            label = f'1 Std Dev: {std_deviation} hours')
plt.axvline(mean_battery_life + std_deviation, color = 'purple', linestyle =':')

```
