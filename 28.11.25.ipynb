```python
a=open("ad.txt",'rt')

am = a.read()
print(am)
a.close()
```

```python
with open('ad.txt') as f:
    print(f.read(19))
```

```python
#readlines
am = open("ad.txt")
print(am.readlines())

am.close()
```

```python
#readline
am = open("ad.txt")
print(am.readline())
print(am.readline())
print(am.readline())

am.close()
```

```python
#write

with open("ad.txt","a") as m: # "a" is used to write in the file
    m.write("life is good") # writes life is good statement in the text file

with open("ad.txt") as m:
    print(m.read())
```

```python
#append "w"

with open("ax.txt","w") as m: # w is used to create a new file even if the file doesn't exist
    m.write("All roads lead to ROME") 

with open("ax.txt") as m:
    print(m.read())
```
