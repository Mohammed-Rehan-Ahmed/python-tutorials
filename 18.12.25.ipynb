```python
import matplotlib.pyplot as mpl
#pyplot is used to do ploting in the output

%matplotlib inline  
#this command is used to clear the cache memory 

subjects = ['Math','Physics','Chemistry']
marks = [72,93,66]

mpl.bar(subjects,marks)
mpl.xlabel('Subjects') #used to label x axis
mpl.ylabel('Marks')    #used to label y axis
mpl.title('Results')   #used to label the title of the chart
mpl.show()
#show 
```

```python
#adding grid to the plot/graph

subjects = ['Math','Physics','Chemistry']
marks = [72,93,66]

mpl.plot(subjects,marks)
mpl.xlabel('Subjects')
mpl.ylabel('Marks')
mpl.title('Results')
mpl.grid(True)
mpl.show()
```

```python
#Scatter plots
subjects = ['Math','Physics','Chemistry']
marks = [72,93,66]

mpl.scatter(subjects,marks)
mpl.xlabel('Subjects')
mpl.ylabel('Marks')
mpl.title('Results')
mpl.grid(True)
mpl.show()
```

```python
#linestyle and color
subjects = ['Math','Physics','Chemistry']
marks = [72,93,66]

mpl.plot(subjects,marks,linestyle='dashed',color = 'cyan') 
# you can also put  '--' instead of dash for the same line style

mpl.xlabel('Subjects')
mpl.ylabel('Marks')
mpl.title('Results')
mpl.show()
```

```python
import numpy as np

x = np.arange(0,10)
y = np.arange(11,21)

x
```

```python
y
```

```python
mpl.scatter(x,y,c = 'purple')  # c  is the short name of color
mpl.xlabel('X - Axis')
mpl.ylabel('Y - Axis')
mpl.title('Scatter')
mpl.show()
```

```python


mpl.plot(x,y,'co',linestyle = 'dashed',linewidth = 2, markersize = 12)  
mpl.xlabel('X - Axis')
mpl.ylabel('Y - Axis')
mpl.title('Plot')
mpl.show()
```

```python
# sub plots

mpl.subplot(2,2,1) #sub
mpl.plot(x,y,'b--')
mpl.subplot(2,2,2)
mpl.plot(y,x,'c*')
mpl.subplot(2,2,3)
mpl.plot(x,y,'ro')
mpl.subplot(2,2,4)
mpl.plot(y,x,'y--')
```

```python
# Home work
math = float(input("Enter math marks"))
phy = float(input("Enter phy marks"))
prg = float(input("Enter programming marks: "))
cmr = float(input("Enter commerce marks : "))

avg = (math + phy + prg + cmr)/4
total = 100
if avg < 50 :
    mpl.bar(avg,total, color = 'red')
else:
    mpl.bar(avg,total, color = 'cyan')
    mpl.show()

```
