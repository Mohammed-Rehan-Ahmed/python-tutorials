```python
import logging
import time

#os stands for operating system, 

logging.basicConfig(
    filename='app.log',
    level=logging.DEBUG,
    format='%(asctime)s-%(levelname)s- %(message)s'
)


# 2. Demo Functions

def start_app():
    logging.info("application started")


def calculate(a,b):
    logging.debug(f'starting caluculation:{a}+{b}')
    return a+b

def check_temperature(temp):
    if temp <30:
        logging.info(f"temperature normal:{temp} C")
    elif 30 <=temp <40:
        logging.warning(f'temperature getting high:{temp} c')
    else:
        logging.error(f"temperature too high:{temp} c")

def critical_failure():
    logging.critical('critical failure syste, shutting down')


# 3. simulate real-time logging

if __name__ == '__main__':
    start_app()

    for i in range(5):
        #Debug /Info logs
        result=calculate(i,i*2)
        logging.info(f'Result:{result}')

        #warning /error logs
        check_temperature(25+ i*5)
        time.sleep(1) # simulate real-time delay
    
    critical_failure() # critical log at the end will display

    logging.info("application ended")
```
