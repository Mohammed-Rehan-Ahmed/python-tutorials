```python
a = []

while True:
    print(''' 1 for append
          2 for pop
          3 for insert
          4 for remove
          5 display''')
    
    b = int(input('enter thre choose'))

    if b == 1:
        c = input("Append a value")
        a.append(c)
        print("Appended",a)
                
    elif b == 2:
        if a:
            remove = a.pop()
            print("popped item:", remove)
        else:
            print("list empty")

    elif b == 3:
        position = int(input("Enter index to insert at: "))
        item = input("Enter item to insert: ")
        a.insert(position, item)
        print("Item inserted!")
    
    elif b == 4:
        item = input("item to remove: ")
        if item in a:
            a.remove(item) 
            print("item removed")

    elif b == 5:
        print("Current list", a)
        
    else:
        print("Exiting")
        break


```
