# Errors

```python
#ZeroDivisionError

x = 10
y = 0

try:
    result = x / y
    print(f"The result is: {result}")
except ZeroDivisionError:
    print("⚠️ Error: Cannot divide by zero. Please use a non-zero denominator.")
    result = None # Set result to a safe value
    
print(f"Execution continues. Result value: {result}")
```

```python
#ValueError

a = input("Enter a whole number: ")

try:
    # Attempt to convert the string input to an integer
    number = int(a)
    print(f"Successfully converted number: {number}")
except ValueError:
    # Handle the case where the input cannot be converted
    print(f"Error: '{a}' is not a valid whole number. Try again.")
    number = 0 # Set number to a safe default value
```

```python
#FileNotFoundError

filename = "my_file.txt"

try:
    # Attempt to open and read the file
    with open(filename, 'r') as file:
        content = file.read()
        print(f"✅ File read successfully. Content snippet: {content[:20]}...")
except FileNotFoundError:
    # Handle the file not found error
    print(f"Error: The file '{filename}' was not found. Check the path.")
    content = "" # Set content to an empty string
except Exception as e:
    # Catch any other potential file errors (e.g., permissions)
    print(f"An unexpected error occurred while handling the file: {e}")
```

```python
try:
    result = 10 / "a"  # Raises TypeError
except TypeError as e:
    # Handles issues with incorrect data types
    print(f"Type Error caught: {e}")
    
except ArithmeticError as e:
    # Handles issues with arithmetic logic (like division by zero)
    print(f"Arithmetic Error caught: {e}")
```
